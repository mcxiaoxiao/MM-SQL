```python
import json
import random

def should_keep(item):
    """判断是否保留该对象"""
    keep = False
    for turn in item['turns']:
        if turn.get('type') in ['ambiguous', 'unanswerable']:
            return True
        if turn.get('type') == 'improper' and len(turn.get('text', '')) > 10:
            return True
    return keep

def process_json_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        data = json.load(file)

    filtered_data = []
    for item in data:
        if should_keep(item):
            filtered_data.append(item)
        else:
            # 对于不满足保留条件的对象，有50%的概率被保留
            if random.random() < 0.3:
                filtered_data.append(item)

    with open(output_file, 'w', encoding='utf-8') as file:
        json.dump(filtered_data, file, ensure_ascii=False, indent=4)

# 调用函数处理文件
process_json_file('MMSQL_test.json', 'filtered_MMSQL_test.json')
```

```python
import json

# 读取JSON文件
with open('filtered_MMSQL_test.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 为每个对象添加id字段
for index, item in enumerate(data):
    item['id'] = index

# 将修改后的数组写回文件
with open('data_with_ids.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=2)

# 输出修改后的JSON字符串到控制台
print(json.dumps(data, ensure_ascii=False, indent=2))
```
