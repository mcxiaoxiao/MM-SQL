```python
import json

def process_turns(turns):
    for i in range(len(turns) - 1):  # 遍历到倒数第二个元素
        current_turn = turns[i]
        if current_turn.get('predict_sql'):  # 检查 predict_sql 是否不为空
            next_turn = current_turn
            predict_text = next_turn.get('predict', '')
            print("Next Turn predict_text:", predict_text)
            
            # 查找 SELECT 和分号的位置
            select_pos = predict_text.upper().find('SELECT')
            colon_pos = predict_text.find(';', select_pos)
            
            if select_pos != -1 and colon_pos != -1:
                predict_sql = predict_text[select_pos:colon_pos + 1].replace('\n', ' ')
            elif select_pos != -1:
                predict_sql = predict_text[select_pos:].replace('\n', ' ')
            else:
                predict_sql = ""
            
            # 将结果存储在新的字段 predict_sql 中
            next_turn['predict_sql'] = predict_sql
            print("new"+predict_sql)

def process_data(data):
    for item in data:
        turns = item.get('turns', [])
        process_turns(turns)


input_filename = 'gpt4_scored_gpt-4-turbo.json'
output_filename = 'edit.json'

# 读取输入文件
with open(input_filename, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 处理数据
process_data(data)

# 将结果写入输出文件
with open(output_filename, 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

```
